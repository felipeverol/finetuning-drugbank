```python
import xml.etree.ElementTree as ET
import json
import random

tree = ET.parse('/kaggle/input/metabolism-absorption-xml/metabolism_absorption.xml')
root = tree.getroot()
print(root)
print(f"Total de drogas: {len(root.findall('drug'))}")

with open('/kaggle/working/metabolism_absorption.jsonl', 'w', encoding='utf-8') as f:
    for drug in root.findall('drug'):
        name = drug.find('name').text
        
        if drug.find('metabolism') != None:
            metabolism = drug.find('metabolism').text
            questions_meta = [
                f"What is the metabolism of {name} like?",
                f"How is {name} metabolized?",
                f"What is the metabolism process of {name}?"
            ]
            user_meta = random.choice(questions_meta)
            entry_meta = {
                "messages": [
                    {"role": "system", "content": 
    """You are an assistant specialized in providing information about the absorption and metabolism of drugs, based on a previously provided database. 
    Only respond with information that exists in that database. 
    Be concise and technical, avoiding speculative or generic answers. 
    If information about the mentioned drug is not available, clearly state that it is unavailable."""},
                    {"role": "user", "content": user_meta},
                    {"role": "assistant", "content": metabolism}
                ]
            }
            f.write(json.dumps(entry_meta, ensure_ascii=False) + '\n')

        
        if drug.find('absorption') != None:
            absorption = drug.find('absorption').text
            questions_abs = [
                f"What is the absorption of {name} like?",
                f"How is {name} absorbed?",
                f"What is the absorption process of {name}?"
            ]
            user_abs = random.choice(questions_abs)
            entry_abs = {
                "messages": [
                    {"role": "system", "content": 
    """You are an assistant specialized in providing information about the absorption and metabolism of drugs, based on a previously provided database. 
    Only respond with information that exists in that database. 
    Be concise and technical, avoiding speculative or generic answers. 
    If information about the mentioned drug is not available, clearly state that it is unavailable."""},
                    {"role": "user", "content": user_abs},
                    {"role": "assistant", "content": absorption}
                ]
            }
            f.write(json.dumps(entry_abs, ensure_ascii=False) + '\n')
```
