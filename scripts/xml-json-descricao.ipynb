```python
import xml.etree.ElementTree as ET
import json
import random

tree = ET.parse('/kaggle/input/drug-description/drug_description.xml')
root = tree.getroot()
print(root)
print(f"Total de drogas: {len(root.findall('drug'))}")

with open('/kaggle/working/drug_interactions.jsonl', 'w', encoding='utf-8') as f:
    for drug in root.findall('drug'):
        name = drug.find('name').text
        description = drug.find('description').text
        
        questions = [
            f"Descreva a droga {name}",
            f"O que você sabe sobre {name}?",
            f"Quais são as características de {name}?"
        ]
        
        user_content = random.choice(questions)
        
        entry = {
            "messages": [
                {"role": "system", "content": "Você é um assistente médico especializado unicamente em responder sobre a descrição de fármacos."},
                {"role": "user", "content": user_content},
                {"role": "assistant", "content": description}
            ]
        }
        f.write(json.dumps(entry, ensure_ascii=False) + '\n')
```
